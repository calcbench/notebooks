```python
import calcbench as cb
%pylab inline
```

```python
tickers = cb.tickers(index='SP500')
```

```python
income_statement_metrics = [
    'revenue', 
    'costofrevenue', 
    'sgaexpense', 
    'operatingexpenses', 
    'operatingincome',
    'ebit',
    'interestexpense',
    'incometaxes',
    'netincome']
```

```python
income_statement_points = cb.normalized_data(entire_universe=True,
                                             metrics=income_statement_metrics,
                                             start_year=2010,
                                             start_period=1,
                                             end_year=2016,
                                             end_period=).T
```

```python
income_statement_points = income_statement_points[income_statement_points.columns[::-1]] #Restore date ordering
```

```python
totals = income_statement_points.sum(level=0)
```

```python
totals = totals.reindex(index=income_statement_metrics) #Restore Ordering
```

```python
totals.T.plot(figsize=(30, 15))
```
