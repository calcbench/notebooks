```python
import calcbench as cb
from calcbench.models.revisions import Revisions
from calcbench.models.period import Period
```

```python
cb.api_client._rig_for_testing('localhost')
```

```python
tickers = cb.tickers(index="DJIA")
```

```python
standardized_metrics = [
    "revenue",
    "assets",
    "IncomeTaxExpense",
    "DeferredLocalIncomeTaxExpense",
]
```

```python
standardized = cb.standardized_raw(
    include_trace=True,
    pit_V2=True,
    company_identifiers=tickers,
    year=2020,
    period=Period.Annual,
    metrics=standardized_metrics,
    point_in_time=True,
    revisions=Revisions.MostRecent,
    XBRL_only=True,
)
```

```python
for (i, t) in enumerate(sorted(standardized)):
    print(
        i,
        t.ticker,
        t.fiscal_year,
        t.fiscal_period,
        t.date_reported,
        t.XBRL,
        t.preliminary,
        t.metric,
        len(t.trace_facts)
      #  ",".join(repr(f) for f in t.trace_facts) if t.trace_facts else "none",
    )
```
