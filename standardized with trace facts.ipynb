```python
import calcbench as cb
from calcbench.models.revisions import Revisions
```

```python
cb.api_client._rig_for_testing("localhost")
```

```python
tickers = cb.tickers(index="DJIA")
```

```python
d = cb.standardized_raw(
    include_trace=True,
    pit_V2=True,
    company_identifiers=tickers,
    all_history=True,
    metrics=["UnrecognizedTaxBenefits"],
    point_in_time=True,
    revisions=Revisions.MostRecent,
    XBRL_only=True,
)
```

```python
for t in d:
    print(
        t.ticker,
        t.fiscal_year,
        t.fiscal_period,
        t.date_reported,
        t.metric,
        ",".join(str(f) for f in t.trace_facts) if t.trace_facts else "none",
    )
```

```python
one_period = cb.standardized_raw(
    include_trace=True,
    pit_V2=True,
    company_identifiers=["DIS"],
    metrics=["Revenue"],
    point_in_time=True,
    revisions=Revisions.MostRecent,
    XBRL_only=True,
    year=2012,
    period=4,
    use_fiscal_period=True,
)

one_period = sorted(one_period, key=lambda d: (d.fiscal_year, d.fiscal_period))

for t in one_period:
    print(
        t.ticker,
        t.fiscal_year,
        t.fiscal_period,
        t.metric,
        ",".join(f.local_name for f in t.trace_facts) if t.trace_facts else "none",
    )
```
