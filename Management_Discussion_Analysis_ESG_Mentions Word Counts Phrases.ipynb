```python
disclosure_contents = pd.read_pickle(contents_pkl_file)
```

```python
matcher = Matcher(nlp.vocab)
matcher.add("management", [[{"LOWER": "management"}]])
```

```python
def disclosure_matches(disclosures):
    documents = nlp.pipe(d[:1000000] for d in disclosures)
    matches = []
    for doc in documents:
        matches.append(matcher(doc))
    return matches
```

```python
phrase_matches = disclosure_contents.progress_apply(disclosure_matches)
```
