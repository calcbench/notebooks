```python
disclosures_to_get = [
    #  "RiskFactors",
    "ManagementsDiscussionAndAnalysis",
]

disclosures = list(
    cb.disclosure_search(
        company_identifiers=["msft"],
        disclosure_names=disclosures_to_get,
        year=2022,
        period_type="annual",
    )
)

contents = disclosures[0].get_disclosure(split_into_paragraphs=True).contents
```

```python
headers = list(categorize_sections.possible_headers(contents))
```

```python
paragraphs = []
for i in range(len(headers)-1):
    paragraph = categorize_sections._html_between_tags(headers[i], headers[i + 1], contents)
    paragraphs.append(paragraph)
```

```python
paragraphs_html = "<h1>paragraph</h1>".join(paragraphs)

html = f"""<div style='width: 100% display: table'><div style='display:table-row'>
                 <div style='width:50%; display: table-cell'>{contents}</div>
                 <div style='display:table-cell'>{paragraphs_html}</div>
             </div>
             """

display(HTML(html))
```
