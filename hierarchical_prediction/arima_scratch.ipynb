```python
import pandas as pd
from statsforecast.core import StatsForecast
from statsforecast.models import AutoARIMA, Naive,AutoETS
```

```python
Y_train_df = pd.read_pickle("company_slash_five.pkl")
```

```python
fcst = StatsForecast(models=[AutoARIMA(season_length=4), Naive(), AutoETS()], freq="QE", n_jobs=-1)
```

```python
Y_hat_df = fcst.forecast(
    df=Y_train_df,
    h=4,
    fitted=True,
)
```

```python
fcst.plot(
    Y_train_df,
    Y_hat_df,
    models=["AutoARIMA", 'AutoETS'],
    unique_ids=["company/5"],
)
```
